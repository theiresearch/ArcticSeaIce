# Arctic Project in Correlation Matrix

## Import libraries

```r
library(corrplot)
```

## Load data

```r
arctic <-read.csv("arctic_data.csv",stringsAsFactors = F)
```

## Correlation matrix

```r
feature_correlations <- cor(arctic[,5:16])
corrplot(feature_correlations,method="circle",type="upper")
corrplot(feature_correlations,method="number",number.cex = 0.5,type="upper")
```
